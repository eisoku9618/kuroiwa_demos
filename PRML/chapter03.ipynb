## 第3章 線形回帰モデル

- http://www.cns.atr.jp/~mfukushima/PRML_resume.pdf

がわかりやすかった．

第1章でやっていた線形回帰モデルは
$$
y(x, \textbf{w}) = w_0 + w_1 x + w_2 x^2 + \ldots + w_M x^M = \sum_{j=0}^M w_j x^j
$$
というものだった．

これを拡張して，基底関数$\phi_j(\textbf{x})$というものを導入して
$$
\begin{align}
y(\textbf{x}, \textbf{w}) &=& w_0 + \sum_{j=1}^{M-1} w_j \phi_j(\textbf{x})\\
&=&\sum_{j=0}^{M-1} w_j \phi_j(\textbf{x})\\
&=&\begin{pmatrix}w_0 w_1 \ldots w_{M-1}
\end{pmatrix}\begin{pmatrix}
\phi_0(\textbf{x})\\
\phi_1(\textbf{x})\\
\vdots\\
\phi_{M-1}(\textbf{x})
\end{pmatrix}\\
&=&\textbf{w}^{T} \textbf{$\phi$} (\textbf{x})
\end{align}
$$
とする（$\phi_0(\textbf{x})=1$）．
非線形な基底関数を使うことで，表現力が豊かになって嬉しい．
パラメータ$\textbf{w}$に関しては線形であることに注意．

基底関数は何でもいいけど，よくあるやつは
- 第1章で使っていたのは$\phi_j(x) = x^j$
- ガウス基底関数$\phi_j(x) = \mathrm{exp}\left\{-\frac{(x-\mu_j)^2}{2s^2}\right\}$
- シグモイド基底関数$\phi_j(x) = \sigma \left(\frac{x-\mu_j}{s}\right) = \frac{1}{1+\mathrm{exp}(-a)} \left(\frac{x-\mu_j}{s}\right)$

### 最尤推定と最小二乗法の関係
目標変数$t$が
$$
t = y(\textbf{x}, \textbf{w})+\epsilon
$$
と与えられる場合を考える．ただし，$\epsilon$は期待値0で分散の逆数（＝精度）が$\beta$のガウス確率変数とする．

そうすると
$$
p(t|\textbf{x}, \textbf{w}, \beta) = \mathcal{N} (t | y(\textbf{x}, \textbf{w}), \beta^{-1})
$$
と表せる．
この２式の関係はよく覚えておくのが良さそう．

入力$\textbf{X} = \left\{\textbf{x}_1, \ldots, \textbf{x}_N\right\}$とそれに対応する目標値$\textbf{t}=\left\{t_1,\ldots, t_N\right\}$として，
これらのデータ点は上の式$p(t|\textbf{x}, \textbf{w}, \beta)$から独立に生成されたとすると，
尤度関数は
$$
p(\textbf{t} | \textbf{X}, \textbf{w}, \beta) = \Pi_{n=1}^{N} \mathcal{N} (t_n | y(\textbf{x}_n, \textbf{w}), \beta^{-1}) = \Pi_{n=1}^{N}\mathcal{N}(t_n | \textbf{w}^{T} \textbf{$\phi$} (\textbf{x}_n), \beta^{-1})
$$
となる．

尤度関数のログをとって
$$
\begin{align}
\ln p(\textbf{t} | \textbf{X}, \textbf{w}, \beta) = \ln p(\textbf{t} | \textbf{w}, \beta) &=& \sum_{n=1}^{N}\ln\mathcal{N}(t_n | \textbf{w}^{T} \textbf{$\phi$} (\textbf{x}_n), \beta^{-1})\\
&=& \sum_{n=1}^{N}\ln \left\{ \left(\frac{1}{2\pi}\right)^{\frac{1}{2}}\beta^{\frac{1}{2}} \exp\left\{-\frac{\beta}{2} \left(t_n - \textbf{w}^{T} \textbf{$\phi$} (\textbf{x}_n)\right)^2\right\} \right\} \\
&=& \frac{N}{2}\ln \beta - \frac{N}{2} \ln(2\pi) - \beta \frac{1}{2} \sum_{n=1}^{N} \left\{ t_n - \textbf{w}^{T} \textbf{$\phi$} (\textbf{x}_n) \right\}^2
\end{align}
$$
これが$\textbf{w}$に関して最大化すれば．最尤推定によって$\textbf{w}$が求まるので
$$
\begin{align}
&& \frac{\partial \ln p(\textbf{t} | \textbf{w}, \beta)}{\partial \textbf{w}}=0\\
&\Leftrightarrow& \beta \sum_{n=1}^{N} \left\{ t_n - \textbf{w}^{T} \textbf{$\phi$} (\textbf{x}_n) \right\} \textbf{$\phi$} (\textbf{x}_n)^{T} = 0\\
&\Leftrightarrow& \sum_{n=1}^{N} t_n \textbf{$\phi$} (\textbf{x}_n)^{T} = \textbf{w}^{T} \sum_{n=1}^{N} \textbf{$\phi$} (\textbf{x}_n) \textbf{$\phi$} (\textbf{x}_n)^{T}\\
&\Leftrightarrow& \sum_{n=1}^{N} t_n \textbf{$\phi$} (\textbf{x}_n) =  \left(\sum_{n=1}^{N} \textbf{$\phi$} (\textbf{x}_n) \textbf{$\phi$} (\textbf{x}_n)^{T} \right)\textbf{w}\\
&\Leftrightarrow&
t_1 \begin{pmatrix} \phi_0(\textbf{x}_1)\\ \phi_1(\textbf{x}_1)\\ \vdots\\ \phi_{M-1}(\textbf{x}_1)
\end{pmatrix} +
t_2 \begin{pmatrix} \phi_0(\textbf{x}_2)\\ \phi_1(\textbf{x}_2)\\ \vdots\\ \phi_{M-1}(\textbf{x}_2)
\end{pmatrix} +
\cdots +
t_N \begin{pmatrix} \phi_0(\textbf{x}_N)\\ \phi_1(\textbf{x}_N)\\ \vdots\\ \phi_{M-1}(\textbf{x}_N)
\end{pmatrix}
=  \left(\sum_{n=1}^{N} \textbf{$\phi$} (\textbf{x}_n) \textbf{$\phi$} (\textbf{x}_n)^{T} \right)\textbf{w}\\
&\Leftrightarrow& \begin{pmatrix}
\phi_0(\textbf{x}_1) & \phi_0(\textbf{x}_2) & \cdots & \phi_0(\textbf{x}_N)\\
\phi_1(\textbf{x}_1) & \phi_1(\textbf{x}_2) & \cdots & \phi_1(\textbf{x}_N)\\
\vdots\\
\phi_{M-1}(\textbf{x}_1) & \phi_{M-1}(\textbf{x}_2) & \cdots & \phi_{M-1}(\textbf{x}_N)
\end{pmatrix}
\begin{pmatrix}
t_1\\
t_2\\
\vdots\\
t_N
\end{pmatrix} =  \left(\sum_{n=1}^{N} \textbf{$\phi$} (\textbf{x}_n) \textbf{$\phi$} (\textbf{x}_n)^{T} \right)\textbf{w}\\
&\Leftrightarrow& \textbf{$\Phi$}^{T} \textbf{t} =  \left(\sum_{n=1}^{N} \textbf{$\phi$} (\textbf{x}_n) \textbf{$\phi$} (\textbf{x}_n)^{T} \right)\textbf{w}\\
&\Leftrightarrow& \textbf{$\Phi$}^{T} \textbf{t} =  \textbf{$\Phi$}^{T} \textbf{$\Phi$} \textbf{w}\\
&\Leftrightarrow& \textbf{w} = \left(\textbf{$\Phi$}^{T} \textbf{$\Phi$}\right)^{-1}\textbf{$\Phi$}^{T} \textbf{t}
\end{align}
$$
また，$\beta$に関して最大化すれば，最尤推定によって$\beta$が求まるので
$$
\begin{align}
&& \frac{\partial \ln p(\textbf{t} | \textbf{w}, \beta)}{\partial \beta}=0\\
&\Leftrightarrow& 
\frac{N}{2}\frac{1}{\beta} = \frac{1}{2} \sum_{n=1}^{N} \left\{ t_n - \textbf{w}^{T} \textbf{$\phi$} (\textbf{x}_n) \right\}^2\\
&\Leftrightarrow& 
\frac{1}{\beta} = \frac{1}{N} \sum_{n=1}^{N} \left\{ t_n - \textbf{w}^{T} \textbf{$\phi$} (\textbf{x}_n) \right\}^2
\end{align}
$$

となる．
これは直感と一致する．

## ベイズ線形回帰

上の話で，$\lambda$が大事だなーということが分かって，これを自動的に決まってくれると嬉しくて，ベイズ的に取り扱うとこれがクリアできるという噂．

### パラメータの分布

モデルパラメータ$\textbf{w}$の事後分布を考える話．

モデルパラメータ$\textbf{w}$の事前確率を
$$
\begin{align}
&&p(\textbf{w}) = \mathcal{N} (\textbf{w} | \textbf{m}_0, \textbf{S}_0)\\
&\Leftrightarrow& p(\textbf{w} | \alpha) = \mathcal{N} (\textbf{w} | \textbf{0}, \alpha^{-1} \textbf{I})
\end{align}
$$
と置いてみる．

尤度関数は
$$
p(\textbf{t} | \textbf{X}, \textbf{w}, \beta) = \Pi_{n=1}^{N} \mathcal{N} (t_n | y(\textbf{x}_n, \textbf{w}), \beta^{-1}) = \Pi_{n=1}^{N}\mathcal{N}(t_n | \textbf{w}^{T} \textbf{$\phi$} (\textbf{x}_n), \beta^{-1})
$$
で，事後分布は事前確率と尤度関数の積に比例するので

$$
\begin{align}
p(\textbf{w} | \textbf{t}) &\propto& \mathcal{N} (\textbf{w} | \textbf{m}_0, \textbf{S}_0) \Pi_{n=1}^{N}\mathcal{N}(t_n | \textbf{w}^{T} \textbf{$\phi$} (\textbf{x}_n), \beta^{-1})\\
&\propto& \mathrm{exp} \left(- \frac{1}{2} (\textbf{w} - \textbf{m}_0)^{T} \textbf{S}_0^{-1} (\textbf{w} - \textbf{m}_0) - \sum_{n=1}^{N}\frac{\beta}{2}(t_n - \textbf{w}^{T} \textbf{$\phi$}(\textbf{x}_n))^2 \right)\\
&\propto& \mathrm{exp} \left(- \frac{1}{2} \left\{(\textbf{w} - \textbf{m}_0)^{T} \textbf{S}_0^{-1} (\textbf{w} - \textbf{m}_0) + \sum_{n=1}^{N}{\beta}(t_n - \textbf{w}^{T} \textbf{$\phi$}(\textbf{x}_n))^2 \right\}\right)\\
&\propto& \mathrm{exp} \left(- \frac{1}{2} \left\{
\textbf{w}^{T} \left( \textbf{S}_0^{-1} + \beta \sum_{n=1}^{N} \textbf{$\phi$}(\textbf{x}_n) \textbf{$\phi$}(\textbf{x}_n)^{T} \right) \textbf{w} - 2 \textbf{w}^{T} \left( \textbf{S}_0^{-1} \textbf{m}_0 + \beta \sum_{n=1}^{N} \textbf{$\phi$}(\textbf{x}_n) t_n \right)  \right\}\right)\\
&\propto& \mathrm{exp} \left(- \frac{1}{2} \left\{
\textbf{w}^{T} \left( \textbf{S}_0^{-1} + \beta \textbf{$\Phi$}^{T} \textbf{$\Phi$} \right) \textbf{w} - 2 \textbf{w}^{T} \left( \textbf{S}_0^{-1} \textbf{m}_0 + \beta \textbf{$\Phi$}^{T} \textbf{t} \right)  \right\}\right)\\
&\propto& \mathcal{N} (\textbf{w} | \textbf{m}_N, \textbf{S}_N)
\end{align}
$$

となる．係数比較すると
$$
\begin{align}
\textbf{S}_N^{-1} &=& \textbf{S}_0^{-1} + \beta \textbf{$\Phi$}^{T} \textbf{$\Phi$}\\
\textbf{m}_N  &=& \textbf{S}_N \left(\textbf{S}_0^{-1} \textbf{m}_0 + \beta \textbf{$\Phi$}^{T} \textbf{t}\right)
\end{align}
$$
となって，$\textbf{m}_0 = \textbf{0}$と$\textbf{S}_0 = \alpha^{-1} \textbf{I}$を代入すると
$$
\begin{align}
\textbf{S}_N^{-1} &=& \alpha \textbf{I} + \beta \textbf{$\Phi$}^{T} \textbf{$\Phi$}\\
\textbf{m}_N  &=& \beta \textbf{S}_N \textbf{$\Phi$}^{T} \textbf{t}
\end{align}
$$

### 予測分布

新しい$x$の値に対する$t$を予測するときの分布をベイズ的に考える話．
上のパラメータの分布のところで求めた事後分布を使って，予測分布を求める．

#### 図3.7の再現

前提条件は
$$
\begin{align}
\alpha &=& 2.0\\
\beta  &=& \left(\frac{1}{0.2}\right)^2 = 25\\
\textbf{$\phi$}(x) &=& \begin{pmatrix} \phi_0(x)\\ \phi_1(x)\end{pmatrix} = \begin{pmatrix} 1\\ x\end{pmatrix}\\
\textbf{$\Phi$} &=& \begin{pmatrix}1 & x\end{pmatrix}
\end{align}
$$
とする．

ただし，
$$
精度パラメータ：\beta\\
標準偏差：\sigma\\
分散：\sigma^2\\
\beta = \frac{1}{\sigma^2}\\
\sigma = \frac{1}{\sqrt{\beta}}\\
\mathcal{N} (x | \mu, \sigma^2) = \mathcal{N} (x | \mu, \beta^{-1}) = \frac{1}{\sqrt{2\pi\sigma^2}}\mathrm{e}^{-\frac{1}{2\sigma^2}(x-\mu)^2}
$$
である．

また，このパターンは「データ点が逐次的に与えられる」場合なので，それを意識する．
式(3.53)(3.54)ではなくて式(3.50)(3.51)を使う，という意味で．

```python
%matplotlib inline
import numpy
import matplotlib.pyplot

# parameter
a0 = -0.3
a1 = 0.5
N = 20
alpha = 2.0
std = 0.2
beta = (1 / std)**2

# training data
xn = numpy.random.uniform(-1, 1, N)
tn = (a0 + xn * a1) + numpy.random.normal(0, std, N)
mu_N = numpy.matrix([0,0]).T
Sigma_N = (1.0 / alpha) * numpy.eye(2)

# basic variable and function
resolition_for_w = 50
w0s = numpy.linspace(-1, +1, resolition_for_w)
w1s = numpy.linspace(-1, +1, resolition_for_w)
W0s, W1s = numpy.meshgrid(w0s, w1s)
xs = numpy.linspace(-1.0, +1.0, 2)
def gaussian_1D(x, mu, sigma):
    return 1.0 / numpy.sqrt(2.0*numpy.pi*sigma**2) * numpy.exp(-1.0 / (2*sigma**2) * (x - mu)**2)
def gaussian_2D(x, mu, sigma):
    tmp = 1.0 / (2.0*numpy.pi * numpy.sqrt(numpy.linalg.det(sigma))) * numpy.exp(-1.0 / 2.0 * (x - mu).T * numpy.linalg.inv(sigma) * (x - mu))
    # tmp is expected as numpy.matrix which size is 1x1
    return tmp[0, 0]

matplotlib.pyplot.rcParams['figure.figsize'] = (15.0, 15.0)
fig, (axes) = matplotlib.pyplot.subplots(4, 3)

for i in range(N):
    if i == 0 or i == 1 or i == 2 or i == N-1:
        # plot middle row
        Z = numpy.array([gaussian_2D(numpy.matrix([w0,w1]).T, mu_N, Sigma_N) for w1 in w1s for w0 in w0s]).reshape(resolition_for_w, resolition_for_w)
        if i == N-1:
            axes[3][1].pcolor(W0s, W1s, Z)
            axes[3][1].scatter(a0, a1, c='white', s=150, marker='+')
        else:
            axes[i][1].pcolor(W0s, W1s, Z)
            axes[i][1].scatter(a0, a1, c='white', s=150, marker='+')

        # plot right row
        for j in range(6):
            # ws is a sampling [w0, w1] set
            ws = numpy.random.multivariate_normal(mu_N.T.tolist()[0], Sigma_N.tolist())
            ys_tmp = [numpy.dot(ws, [1, x]) for x in xs]
            if i == N-1:
                axes[3][2].plot(xs, ys_tmp)
            else:
                axes[i][2].plot(xs, ys_tmp)
        if i != 0:
            if i == N-1:
                axes[3][2].scatter(xn[:i], tn[:i], c='blue', facecolors='none', s=50)
            else:
                axes[i][2].scatter(xn[:i], tn[:i], c='blue', facecolors='none', s=50)

        # plot left row
        Z = numpy.array([gaussian_1D(tn[i], w0+w1*xn[i], std) for w1 in w1s for w0 in w0s]).reshape(resolition_for_w, resolition_for_w)
        if i == N-1:
            pass
        else:
            axes[i+1][0].pcolor(W0s, W1s, Z)
            axes[i+1][0].scatter(a0, a1, c='white', s=150, marker='+')

    # proc one step
    prev_Sigma_inv = numpy.linalg.inv(Sigma_N)
    # eq. (3.51)
    new_Sigma_inv = prev_Sigma_inv + beta * numpy.array([[1, xn[i]],[xn[i], xn[i]**2]])
    Sigma_N = numpy.linalg.inv(new_Sigma_inv)
    # eq. (3.50)
    mu_N = Sigma_N * (prev_Sigma_inv * mu_N + beta * numpy.matrix([1, xn[i]]).T * tn[i])

# set limit for x-axis and y-axis
rangelimiter = numpy.vectorize(lambda ax: ax.set_xlim(-1.0, +1.0))
rangelimiter(axes)
rangelimiter = numpy.vectorize(lambda ax: ax.set_ylim(-1.0, +1.0))
rangelimiter(axes)

matplotlib.pyplot.show()
```

左側の図が直線っぽくなる理由は，尤度関数の指数部$t_n - (w_0 + w_1 x_n)$が0になる部分が真っ黒になるはずだから．

また，二行目左の直線が，青丸付近を通っているのがポイントで，これは$\textbf{w}$の事後分布が良い感じに更新されているということ．

#### 図3.8/3.9の再現

まず，加法定理・乗法定理より，
$$
p(\textbf{y}) = \int p(\textbf{x}, \textbf{y}) d\textbf{x} = \int p(\textbf{y} | \textbf{x}) p(\textbf{x}) d\textbf{x}
$$
であり，
$$
p(\textbf{x}) = \mathcal{N}(\textbf{x} | \textbf{$\mu$}, \textbf{$\Lambda$}^{-1})
$$
$$
p(\textbf{y} | \textbf{x}) = \mathcal{N}(\textbf{y} | \textbf{A} \textbf{x} + \textbf{b}, \textbf{L}^{-1})
$$
と書けるときには，第2章の式(2.115)らへんの議論の結果より，
$$
p(\textbf{y}) = \mathcal{N}(\textbf{y} | \textbf{A} \textbf{$\mu$} + \textbf{b}, \textbf{L}^{-1} + \textbf{A} \textbf{$\Lambda$}^{-1} \textbf{A}^{T})
$$
となるらしい．

この公式を使う．

新しい$x$が与えられた時の予測分布は
$$
\begin{align}
p(t | \textbf{t}, \alpha, \beta) &=& \int p(t|\textbf{w}, \beta) p(\textbf{w} | \textbf{t}, \alpha, \beta) d\textbf{w}\\
&=& \int \mathcal{N} (t | \textbf{$\phi$} (\textbf{x})^{T} \textbf{w}, \beta^{-1}) \mathcal{N} (\textbf{w} | \textbf{m}_N, \textbf{S}_N)  d\textbf{w}\\
&=& \mathcal{N} (t | \textbf{$\phi$} (\textbf{x})^{T} \textbf{m}_N, \beta^{-1} + \textbf{$\phi$} (\textbf{x})^{T} \textbf{S}_N \textbf{$\phi$} (\textbf{x}))\\
&=& \mathcal{N} (t | \textbf{m}_N^{T} \textbf{$\phi$} (\textbf{x}), \beta^{-1} + \textbf{$\phi$} (\textbf{x})^{T} \textbf{S}_N \textbf{$\phi$} (\textbf{x}))\\
\end{align}
$$
となる．

```python
%matplotlib inline
import numpy
import matplotlib.pyplot
import random

# parameters
s = 0.1
alpha = 0.1
beta = 9
resolution_for_x = 50
M = 9
N_list = [1, 2, 4, 16, 25, 50]

# training data
xn_all = numpy.linspace(0, 1, resolution_for_x)
tn_all = numpy.sin(xn_all*numpy.pi*2) + numpy.random.normal(loc=0.0, scale=numpy.sqrt(1.0/beta), size=resolution_for_x)
random_order = range(resolution_for_x)
random.shuffle(random_order)

# basic functions
def gaussian_basis_func(s, mu):
    """
    gauss basis function
    eq. (3.4)
    """
    return lambda x:numpy.exp(-(x-mu)**2 / (2*s**2))
base_list = [gaussian_basis_func(s, mu) for mu in numpy.linspace(0, 1, M)]

def calc_design_matrix(xn):
    """
    eq. (3.16)
    """
    ret = numpy.array([])
    for x in xn:
        ret = numpy.append(ret, [base(x) for base in base_list])
    return numpy.reshape(ret, (numpy.size(xn), M))

def calc_predictive_distribution(xn, tn):
    """
    eq. (3.58)
    x in eq. 3.58 is not training data but new input data.
    We should pay attention to the difference because this is a hamari point.
    """
    PHI = calc_design_matrix(xn)
    S_N_inv = alpha * numpy.eye(M) + beta * numpy.dot(PHI.T, PHI)
    S_N = numpy.linalg.inv(S_N_inv)
    m_N = beta * numpy.dot(S_N, numpy.dot(PHI.T, tn))
    def func(x):
        phi_x = [base(x) for base in base_list]
        mu = numpy.dot(m_N.T, phi_x)
        # var = 1 / beta + phi^T S_N phi
        # phi_x is not vector but list in this sentence, so transpose is not necessary.
        var_N = 1.0 / beta + numpy.dot(phi_x, numpy.dot(S_N, phi_x))
        return (mu, var_N)
    return m_N, S_N, func

matplotlib.pyplot.rcParams['figure.figsize'] = (15.0, 15.0)
fig, (axes) = matplotlib.pyplot.subplots(len(N_list), 2)


# plot ground truth
plot_sin = numpy.vectorize(lambda ax: ax.plot(xn_all, numpy.sin(xn_all*numpy.pi*2), color='black'))
plot_sin(axes)

# plot mean and std
for i, N in enumerate(N_list):
    order = random_order[:N]
    xn = [xn_all[idx] for idx in order]
    tn = [tn_all[idx] for idx in order]
    m_N, S_N, f = calc_predictive_distribution(xn, tn)
    y_h = []
    y_m = []
    y_l = []
    for mu, var_N in [f(x) for x in xn_all]:
        std = numpy.sqrt(var_N)
        y_h.append(mu + std)
        y_m.append(mu)
        y_l.append(mu - std)
    # fig. 3.8
    axes[i][0].plot(xn_all, y_m, ':')
    axes[i][0].fill_between(xn_all, y_h, y_l, color='gray')
    axes[i][0].scatter(xn, tn, c='black', facecolors='none', s=50)
    axes[i][1].scatter(xn, tn, c='black', facecolors='none', s=50)
    # fig. 3.9
    for j in range(5):
        w = numpy.random.multivariate_normal(m_N, S_N)
        y = numpy.dot(w, [base(xn_all) for base in base_list])
        axes[i][1].plot(xn_all, y, ':')

# set limit for x-axis and y-axis
rangelimiter = numpy.vectorize(lambda ax: ax.set_xlim(0, +1.0))
rangelimiter(axes)
rangelimiter = numpy.vectorize(lambda ax: ax.set_ylim(-1.5, +1.5))
rangelimiter(axes)

matplotlib.pyplot.show()
```
